```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_df = sns.load_dataset("titanic")
titanic_df.info()
```

```python
# 상관 행렬 계산
# 수치형 데이터만 선택
numeric_df = titanic_df.select_dtypes(include=["Float64", "int64"])

# 상관 행렬 계산
matrix = numeric_df.corr()

# heatmap
sns.heatmap(matrix, vmin=-1.0, vmax=1.0, annot=True)
# annot는 그래프에 값 나오게 하는 것
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("movie_ratings.csv")

#파이 차트의 값
values = df["Genre"].value_counts()

#파이 차트의 라벨
labels = values.index

plt.pie(x=values, labels=labels, autopct="%1.1f%%")
plt.show()
```

```python
#실습
import seaborn as sns
import matplotlib.pyplot as plt

exercise_df = sns.load_dataset("exercise")
exercise_df.info()
exercise_df.head()
```

```python
sns.barplot(data=exercise_df, x="time", y="pulse")
plt.show()
```

```python
sns.barplot(data=exercise_df, x="diet", y="pulse")
plt.show()
```

```python
sns.countplot(data=exercise_df, x="diet")
plt.show()
```

```python
values = exercise_df["diet"].value_counts()
labels = values.index
plt.pie(values, labels=labels, autopct="%1.1f%%")
plt.show()
```
