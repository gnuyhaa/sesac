```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins_df = sns.load_dataset("penguins")
penguins_df.head()
```

```python
sns.histplot(data = penguins_df,
            x = "bill_length_mm",
            bins = 15, # 히스토그램 막대 개수
            )
plt.show()
```

```python
sns.histplot(
    penguins_df,
    x="bill_length_mm",
    hue="species",
    multiple="layer",
    bins=15,
)
plt.show()
# multiple
# layer : 기본값, 겹쳐서 표현
# stack : 쌓아서 구분해서 표현
# dodge : 옆으로 나란히 구분해서 표현 
# fill : 각 범주의 막대 높이를 비율로 표현 
```

```python
# 펭귄의 종에 따른 날개 길이 분포 
sns.kdeplot(data=penguins_df,
            x = 'flipper_length_mm',
           hue = 'species',
           fill=True,
           bw_adjust=10)
plt.show()
```

```python
sns.rugplot(
    data=penguins_df,
    x="flipper_length_mm",
    hue="species",
)

plt.show()
```

```python
sns.displot(
    data=penguins_df,
    x="body_mass_g",
    hue="species",
    kind = 'kde',
    rug = True,
    
)

plt.show()
```
