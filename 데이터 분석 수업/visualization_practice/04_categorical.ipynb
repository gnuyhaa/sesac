```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_df = sns.load_dataset("iris")
iris_df.info()
# sepal : 꽃받침 길이 / 너비
# petal : 꽃잎 길이 / 너비
```

```python
sns.barplot(data=iris_df, x="species", y="sepal_length")
plt.show()
```

```python
sns.barplot(
    data=iris_df, 
    x="species", 
    y="sepal_length", 
    estimator="sum")
plt.show()
#적용할 집계 함수 estimator
```
