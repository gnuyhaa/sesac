```python
# tips 데이터를 활용한 산점도
# 모듈 불러오기 
import seaborn as sns
import matplotlib.pyplot as plt

tips_df = sns.load_dataset("tips")
```

```python
tips_df.info()
```

```python
# 산점도
# 2개의 연속형 데이터를 기반으로 

# total_bill에 따른 tip을 산점도로 시각화
sns.scatterplot(data=tips_df, 
               x = 'total_bill',
               y = 'tip')

plt.show()
```

```python
# hue 매개변수
# 점(포인트)를 구분하는 범주형 데이터
sns.scatterplot(data=tips_df, 
               x = 'total_bill',
               y = 'tip',
               hue='size')

plt.show()
```
