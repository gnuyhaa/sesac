```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# seaborn 실습 데이터 tips
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
tips_df['time'].unique()
```

```python
# 시각화 요소
# 데이터, x축, y축, 구분 데이터(hue)
sns.scatterplot(
    data = tips_df,
    x="total_bill",
    y = "tip",
    hue = 'time'
)
plt.show()
```

```python
# 항공기 탑승객 수
flight_df = sns.load_dataset('flights')
flight_df.head()
```

```python
# lineplot(data,x,y)
sns.lineplot(data=flight_df, x='year',y='passengers')
plt.show()
```

```python
# 월별 승객수
sns.lineplot(data=flight_df, x='month', y='passengers')
plt.show()
```

```python
# facetGrid
# Grid : 격자 형태로 여러개 그래프(서브 플롯) 시각화해서 관계를 확인한다.
tips_df = sns.load_dataset("tips")
```

```python
grid = sns.FacetGrid(
    data = tips_df,
    col = 'day',
)
grid.map(plt.hist, "total_bill")
plt.show()
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.scatterplot(data=titanic, x='age',y='fare')
plt.show()
```

```python
# 시각화 요소
# 데이터, x축, y축, 구분 데이터(hue)
sns.scatterplot(
    data = titanic,
    x="age",
    y = "fare",
    hue = 'survived'
)
plt.show()
```

```python
import seaborn as sns


diamonds = sns.load_dataset("diamonds")

print(diamonds.head())
```

```python
sns.scatterplot(data=diamonds, x='price', y='carat')
plt.show()
```

```python
sns.scatterplot(data=diamonds, x='price', y='carat', hue='color')
plt.show()
```

```python
sns.scatterplot(data=diamonds, x='price', y='carat', hue='cut')
plt.show()
```

```python
grid = sns.FacetGrid(data=diamonds, col="cut")

grid.map(sns.scatterplot, "price", "carat")

plt.show()
```
