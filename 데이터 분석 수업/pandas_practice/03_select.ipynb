```python
# 1. selling_price 열 기준으로 내림차순으로 정렬해서 
# 2. 결과를 새로운 변수(sorted_df)에 저장하세요
sorted_df = df.sort_values(by= "selling_price", ascending=False)
sorted_df
```

```python
# 3. sorted_df에 대해서 인덱스 기반 선택을 하세요
# .iloc[0]
sorted_df.iloc[0]
```

```python
# 4. sorted_df에 대해서 레이블 기반 선택을 하세요 
# .loc[0]
sorted_df.loc[0]
```

```python
# 조건식
# 열을 선택해서 해당 열의 개별 데이터와 값을 비교하는 식
# 판매 가격이 (selling_price) 100000 보다 작은 값을 찾아라 

# 1. 열 선택
# 2. 조건 값을 작성
# 3. 비교연산자를 작성
condition = df['selling_price'] < 100000

# 4. loc를 사용해서 데이터를 선택(필터링)
df.loc[condition]
```

```python
# 주행거리(km_driven)이 100000 보다 큰 데이터를 필터링 
driven = df['km_driven'] > 100000
df.loc[driven]
# 기타. name열과 selling_price 선택 
df.loc[driven,['name','selling_price']]
```

```python
# 연료(fuel)가 'Diesel'인 데이터 필터링
Desel = df['fuel'] == "Diesel"
df.loc[Desel]
```

```python
# 판매 유형이 (seller_type)이 "Dealer" 가 아닌 데이터를 필터링
seller = df['seller_type'] != "Dealer"
df.loc[seller]
```

```python
# 주행거리가 100000 보다 작은
condition1 = df["km_driven"] < 100000

# 판매가격이 10000보다 작은
condition2 = df['selling_price'] < 100000

# 두 조건을 모두 만족하는 데이터를 선택 
df.loc[condition1& condition2]
```

```python
# 주행거리가 100000 보다 작은
condition1 = df["km_driven"] < 100000

# 판매가격이 10000보다 작은
condition2 = df['selling_price'] < 100000

# 운전 방식이 (transmission)이 Automatic 인 차량 
condition3 = df['transmission'] == "Automatic"

# 두 조건을 모두 만족하는 데이터를 선택 
df.loc[condition1& condition2 & condition3]
```

```python
# fuel가 Electric 이거나 transmission이 Automatic인 차량을 선택 

condition1 = df['fuel'] == "Electric"
condition3 = df['transmission'] == "Automatic"

df.loc[condition1 | condition3]
```

```python
# year 가 2018보다 크면서
# fuel가 Petrol 또는 Diesel인 데이터를 선택 

year = df["year"] > 2018
fuel1 = df["fuel"] == "Petrol"
fuel2 = df["fuel"] == "Diesel"

df.loc[year & (fuel1 |fuel2)]
```

```python
# 연료(fuel)가 Petrol,Diesel,LPG인 데이터를 탐색
# 같은 열의 여러개의 값을 찾고 싶을때 isin사용 
condition = df['fuel'].isin(['Petrol','Diesel','LPG'])
df.loc[condition]
```

```python
# 연료(fuel)가 Petrol,Diesel,LPG가 아닌 데이터를 탐색
# ~를 통해 False인 데이터를 선택 
df.loc[~condition]
```

```python
# year가 2020, 2018, 2016, 2014인 데이터를 선택 
condition = df['year'].isin([2020,2018,2016,2014])
df.loc[condition]
```
