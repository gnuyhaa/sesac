```python
df.columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']
```

```python
# 열 이름을 수정할 수 있다.
# 다만, 리스트 길이와 기존 열 개수와 동일해야한다.
# df.columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
#        'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
#        '산도', 'sulphates', 'alcohol', '퀄리티']
```

```python
df.head()
```

```python
# 산도를 기준으로 산도를 오름차순 정렬
# by : 어떤 열을 기준으로 정렬?
# ascending : 오른차순 
df.sort_values(by = "pH", ascending=True)
```

```python
# quality를 기준으로 내림차순으로 정렬
df.sort_values(by = "quality", ascending=False)
```

```python
# 주의사항
# 정렬한 결과가 원본 데이터프레임에 적용 x
# 기존 데이터프레임은 변화가 없다. 
# 새로운 데이터프레임을 생성하는 경우!
# 새로운 데이터프레임을 새로운 변수에 저장을 해서 사용해야한다. 
sorted_df = df.sort_values(by="quality", ascending = True)
sorted_df
```

```python
# quality, alcohol 기준 오름차순 정렬 
sorted_df = df.sort_values(by = ["quality", "alcohol"])
sorted_df
```

```python
# quality, alcohol, residual sugar 기준 오름차순 정렬 
sorted_df = df.sort_values(by = ["residual sugar","quality", "alcohol"])
sorted_df
```

```python
# quality 기준 내림차순
# alcohol 기준 오름차순
# residual sugar 기준 오름차순 정렬 코드 작성
sorted_df = df.sort_values(by = ["quality","alcohol","residual sugar"], ascending =[False, True, True])
sorted_df
```

```python
sorted_df.sort_index(ascending = False)
```

```python
# sort_values(by = []) : 열의 값을 기준으로 정렬 
# sort_index() : 행의 이름(인덱스, index)을 기준으로 정렬

# 행의 이름은 인덱스(index)
# 열의 이름은 열의이름, 컬럼 
```
