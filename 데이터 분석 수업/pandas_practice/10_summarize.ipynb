```python
# 부서의 인원 수가 2명보다 많은 부서를 필터링
# 1. 그룹화
grouped = df.groupby('부서')

# 조건 함수
# 매개변수 : 그룹화한 데이터의 개별 행
def greater_count_two(group):
    count = group['이름'].count()
    # count가 2보다 크면 True 2보다 작으면 False
    if count > 2:
        return True
    elif count <=2:
        return False
    
# 2. 필터링
grouped.filter(greater_count_two)
```

```python
# 평균 연봉이 6000 이상인 부서만 필터링 
# 1. 그룹화
grouped = df.groupby("부서")

# 2. 조건 함수 작성 
def greater_salary(group):
    return group['연봉'].mean() >= 6000

# 3. filter() 적용
df.groupby('부서').filter(greater_salary)
```

```python
# 3. 장르별 전 세계 판매량 합계
df.groupby('genre')['total_sales'].sum()
```

```python
# 4. 플랫폼별 세계 판매량 평균
df.groupby('console')['total_sales'].mean()
```

```python
# 5. 개발사별 게임 출시 개수 
df.groupby('developer')['title'].count().sort_values(ascending=False)
```

```python
# 6. 플랫폼별 세계 판매량, 북미 일본, 유럽 및 아프리카, 기타 판매량의 합계와 평균
grouped = df.groupby('console')

result = grouped.agg(
{
    "total_sales":['sum','mean'],
    "na_sales":['sum','mean'],
    "jp_sales":['sum','mean'],
    "pal_sales":['sum','mean'],
    "other_sales":['sum','mean'],
    
}).sort_values(by=('jp_sales','sum'),ascending=False)

# jp_sales 열과 sum 열은 계층구조이기 때문에
# 소괄호 ()를 사용해서 계층열을 표현한다.
result
```

```python
# 7. 평론가 점수의 평균이 9.5가 넘는 게임 개발사 출력 
grouped = df.groupby('developer')

def greater_mean(group):
    return group['critic_score'].mean() > 9.5


df.groupby('developer').filter(greater_mean)
```
