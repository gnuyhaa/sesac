```python
import pandas as pd
import numpy as np
import seaborn as snsfrom
from scipy import stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/Sleep_health_and_lifestyle_dataset.csv")
df.info()
df.head()
```

```python
# 수면장애 여부 컬럼 생성
df["Disease"] = df["Sleep Disorder"].apply(lambda x: 0 if pd.isna(x) else 1)
new_df = df[
    [
        "Age",
        "Sleep Duration",
        "Quality of Sleep",
        "Physical Activity Level",
        "Stress Level",
        "Heart Rate",
        "Daily Steps",
        "Disease",
    ]
]
new_df
```

```python
# 각 컬럼 별 상관계수 산출
df_corr = new_df.corr()
df_corr
```
